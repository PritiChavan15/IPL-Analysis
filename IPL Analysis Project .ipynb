### Find Team wise runs scored.


```python
df.groupby("batting_team")["total_runs"].sum()
```

### Find Total Runs of Virat Kohli


```python
df[df["batsman"]=="V Kohli"]["batsman_runs"].sum()
```

### Find Total Runs of Your Fav Player.


```python
df[df["batsman"]=="RG Sharma"]["batsman_runs"].sum()
```

### Extra Runs Given By All Bowlers in a League.


```python
df.groupby("bowler")["extra_runs"].sum().sort_values(ascending=False)
```

### Extra Runs Given By TS Mills.


```python
df[df["bowler"]=="TS Mills"]["extra_runs"].sum()
```

### Catches Taken By Mandeep Singh.


```python
df[(df.dismissal_kind=="caught")&(df.fielder=="Mandeep Singh")].shape[0]
```
